```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
MU = 1

for cost in range(7,11):
    theta = lambda v: (.75*cost + 3*v**2/(32*cost))/MU
    v = np.linspace(4*cost/3,8*cost/3,10)
    plt.plot(v, v - theta(v) - 1.5*cost,label='lambda = %s'%(cost))

print(v)

plt.legend()
plt.xlabel('v ')
plt.ylabel('v - theta(v)')
plt.savefig('sweet.png')
```
