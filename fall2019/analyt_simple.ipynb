```python
from sympy import *
```

```python
#general purpose code for interior cases

theta1=Symbol('\\theta_1')
theta2 =Symbol('\\theta_2')
lamb =Symbol('\lambda')
mu1 =Symbol('\mu_1')
mu2 =Symbol('\mu_2')
out = Symbol('\pi')

def profit(theta1,theta2, mu1, mu2, lamb):
    return ((3*lamb)+ mu2*theta2 - mu1*theta1)**2/(18*lamb)

def t(theta1, theta2, mu1, mu2, lamb):
    return (3*lamb + mu2*theta2 - mu1*theta1)/(6*lamb)

#analyze symmetric simultaneous case first
def nb_objective(theta1,theta2, mu1, mu2, lamb, outside):
    term1 = (profit(theta1,theta2, mu1, mu2, lamb))
    term2 = ( mu2*theta2 + (mu1*theta1- mu2*theta2)*t(theta1, theta2, mu1, mu2,lamb))
    return term1*(term2 - outside)
```

```python
print(nb_objective(theta1,theta2, mu1, mu2, lamb, out))
```

```python
plot(nb_objective(theta1,1.5*10, .5, .5, 10,0), (theta1,10,70))
```

```python
#working out simultaneous solution
foc =  diff(nb_objective(theta2, theta1,  mu2, mu1, lamb, out), theta2)
foc2 = simplify(lamb**2*12*6*3*foc/((2*mu2*(3*lamb - mu2*theta2 + mu1*theta1) )))
h =solve(foc2,theta2)[0]
print(foc2,'\n')
print(h, '\n')

foc_firm1 =  diff(nb_objective(theta1, theta2,  mu1, mu2, lamb, out), theta1)
foc_firm1_2 = simplify(lamb**2*12*6*3*foc_firm1/((2*mu1*(3*lamb - mu2*theta1 + mu1*theta2) )))
foc_firm1_2 = foc_firm1_2.subs({theta2:h})

print(foc_firm1_2,'\n')

print(simplify(foc_firm1_2),'\n')

print(solve(foc_firm1_2,theta1),'\n')

print(solve(foc.subs({mu1:mu2, theta2:theta1}),theta1))
```
